```python
from math import radians, degrees, cos, sqrt

def get_bounding_box(lat, lon, distance_km=100):
    # Earth radius in kilometers
    R = 6371.0

    # Convert latitude and longitude from degrees to radians
    lat_rad, lon_rad = radians(lat), radians(lon)
    
    # Angular distance in radians
    angular_distance = distance_km / R

    # Calculate min and max latitude
    min_lat = lat - degrees(angular_distance)
    max_lat = lat + degrees(angular_distance)

    # Calculate min and max longitude (adjust for latitude)
    delta_lon = degrees(angular_distance / cos(lat_rad))
    min_lon = lon - delta_lon
    max_lon = lon + delta_lon

    return {
        'min_lat': min_lat,
        'max_lat': max_lat,
        'min_lon': min_lon,
        'max_lon': max_lon
    }

# Example usage
center_lat, center_lon = 50.0875, 14.421389
bbox = get_bounding_box(center_lat, center_lon, 100)
print(bbox)   
```

```python
from geopy.distance import distance

input_point = (50.0875, 14.421389)  # Example reference point
points = [[50.150874, 14.563832], [50.104217, 14.4757]]  # List of points
points_filtered = []

for point in points:
    if distance(input_point, point).km < 100:
        points_filtered.append(point)

print(points_filtered)
```
